```python
cols = ['w_ace', 'w_df',
       'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon', 'w_SvGms', 'w_bpSaved',
       'w_bpFaced', 'l_ace', 'l_df', 'l_svpt', 'l_1stIn', 'l_1stWon',
       'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced']

def chk_df(l, funct, opr):
# this function will get any list, and function and operation
# then create Bitwise Operators for within the list
    a = ''
    for x in range(len(l)):
        if x == len(l)-1:
            a = (a + f'df[\'{l[x]}\'].{funct}')
        else:
            a = (a + f'df[\'{l[x]}\'].{funct} {opr} ')

    a = f'df[{a}]'
    print(a)

chk_df(cols, 'isnull()', '|')
```
