```python
import pandas as pd
```

```python
df = pd.read_csv('./data/challengers/all.csv')
```

```python
df.drop(columns='Unnamed: 0', axis=1, inplace=True)
```

```python
df['winner_name'] = df['winner_name'].str.strip()
```

```python
df.head()
```

```python
df.columns
```

```python
df['tourney_name'].value_counts()
```

```python
df.drop(columns=['tourney_id','draw_size', 'tourney_level', 'tourney_date', 'winner_seed', 'winner_entry', 'match_num', 'winner_hand', 'winner_ht', 'winner_ioc', 'winner_age', 'loser_id', 'loser_seed', 'loser_entry', 'loser_name', 'loser_hand',
       'loser_ht', 'loser_ioc', 'loser_age', 'score', 'best_of', 'round',
       'minutes', 'w_ace', 'w_df', 'w_svpt', 'w_1stIn', 'w_1stWon', 'w_2ndWon',
       'w_SvGms', 'w_bpSaved', 'w_bpFaced', 'l_ace', 'l_df', 'l_svpt',
       'l_1stIn', 'l_1stWon', 'l_2ndWon', 'l_SvGms', 'l_bpSaved', 'l_bpFaced',
       'winner_rank', 'winner_rank_points', 'loser_rank', 'loser_rank_points'], axis=1, inplace=True)
```

```python
df_dummies = pd.get_dummies(df, columns=['tourney_name', 'surface'])
```

```python
df_dummies = df_dummies.sort_values('winner_name')
```

```python
df_dummies.columns
```

```python
df_dummies.columns.unique()
```

```python
x = {}
for i in df_dummies.columns:
    if i != 'winner_name' and i != 'winner_id':
        x[i] = ['sum','mean']

df_dummies = df_dummies.groupby('winner_name').agg(x)
```

```python
df_dummies.head()
```

```python
# split columns name sum/mean, into 2 separate columns

df_final = pd.DataFrame()

for x, y in df_dummies.columns:
    df_final[x+" "+y] = df_dummies[x][y]
```

```python
df_final.columns
```

```python
# rename columns

for c in df_final.columns:
    if 'tourney_name_' in c:
        c2 = c[13:]
        df_final = df_final.rename(columns={c:c2})
```

```python
df_final.columns
```

```python
for x, y in df_dummies.columns:
    total = 0
    avg = 0
    if y == 'sum':
        total += df_dummies[x][y]
    else:
        avg = (df_dummies[x][y]+avg)/2
    
    df_dummies['total_win'] = total
    df_dummies['average_win'] = avg
```

```python
df_final
```

```python
df_final.to_csv('./data/challengers/win.csv')
```
